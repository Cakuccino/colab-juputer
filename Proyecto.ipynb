# Celda 2: Funciones <br>
En esta celda crearemos las funciones que encapsulan la lógica para crear un nuevo registro y agregarlo a la lista principal.


```python
#=================
# CELDA 2: función para registrar nuevos alumnos
# (esta función registraría nuevos productos)
# Para iniciar una función usamos la palabra (def)
def registra_alumno(inventario):
    """
    DocString: Registra un nuevo alumno en el inventario. Implementando validación de entradas básicas.
    """
    print("--Registro de un nuevo Alumno")
    while True:
        codigo = input("Ingrese el código del Alumno (ej, IND001)").strip().upper() 
        if codigo: 
            # Verificar si existen duplicados (validar si un cliente ya está registrado)
            if any(alumno['codigo'] == codigo for alumno in inventario):
                print("Error: El alumno ya fue registrado")
                continue # Revisa todos los datos
            break # rompe la ejecución del código para evitar un ciclo
        print("El código no puede estar vacio")
    # Ingresar datos
    nombre = input("Ingrese el nombre del alumno").strip()
    materia = input("Ingresa la materia del alumno").strip()

    # Agregar las calificaciónes del alumno
    while True:
        try:
            calificacion_str = input("Ingresa la calificación final del alumno")
            calificacion = float(calificacion_str)
            # Validar que la calificación esté dentro del rango de 0 a 10
            if 0 <= calificacion <= 10:
                break
            else:
                print("Error: La calificación debe estar entre 0 y 10")
        except ValueError:
            print("Error: Ingresa un número valido")
# Crear Diccionario que nos permite agregar los datos (cambiar por datos del cliente)
        nuevo_alumno = {
    'codigo': codigo,
    'nombre': nombre,
    'calificacion': calificacion,
    'materia': materia
}

# Agregar alumno al diccionario
        inventario.append(nuevo_alumno)
    print(f"Alumno {nombre} registrado con éxito")

```

## Celda 3: Reporte de alumno y cálculos <br>

CELDA 3: Esta función agrupa los datos de los alumnos para realizar el cálculo del promedio de califcaciones y el resumen de los daots registrados.

```python
# Crear la función para visualizar el reporte de materias
def reporte_materias(inventario):
    """
    Calcula y muestra el promedio de calificaciones por materia registrada.
    (Es similar al cálculo de total de piezas existentes.)
    """
    print("----REPORTE DE CALIFICACIONES POR MATERIA----")
    # CREAR una lista vacía para almacenar las suma de calificaciones
    reporte_agregado = ()
    # Ciclo FOR que va agregando los datos
    for alumno in inventario:
        materia = alumno['materia']
        calificacion = alumno['calificacion']
        # Validaremos que los datos están siendo ingresados correctamente
        if materia not in reporte_agregado:
            reporte_agregado[materia] = { 'suma': 0, 'conteo': 0 }
           # Se irán agregando las materias y calificaciones
        reporte_agregado[materia]['suma'] += calificacion
        reporte_agregado[materia]['conteo'] += 1

    # Se realiza el cálculo de promedios y su visualización
    if not reporte_agregado:
        print("El inventario está vacío.")
        return
    # Realizar el cálculo del promedio
    for materia, datos in reporte_agregado.items():
        promedio = datos['suma'] / datos['conteo']
        print(f"- {materia}: {datos('conteo')} alumnos - Promedio: {promedio:.2f}")
```

## Celda 4: Función de Modificación<br>
Esta función (modificar_calificacion) simula la modificación de existencias. En este caso, cambia la calificación del alumno.

```python
# 4. Función para modificar la calificación (Simula 'Modificar la cantidad de un producto')
def modificar_calificacion(inventario):
    """Permite cambiar la calificación de un alumno existente."""

    codigo_a_modificar = input("Ingrese el código del alumno cuya calificación desea modificar: ").strip().upper()
    
    # Buscar el alumno usando el código
    for alumno in inventario:
        if alumno['codigo'] == codigo_a_modificar:
            print (f"\nAlumno encontrado: {alumno['nombre']}. Calificación actual: {alumno['calificacion']}")

            # Validación de Entradas para la nueva calificación
            while True:
                try:
                    nueva_calificacion_str = input("Ingrese la NUEVA calificación final (0-100): ")
                    nueva_calificacion = float(nueva_calificacion_str)
                    if 0 <= nueva_calificacion <= 100:
                        break
                    else:
                        print("ERROR: La calificación debe estar entre 0 y 100.")
                except ValueError:
                    print("ERROR: Ingrese un número válido.")

            # Actualización del valor en el diccionario
            alumno['calificacion'] = nueva_calificacion
            print(f"\nCalificación de {alumno['nombre']} actualizada a {nueva_calificacion}.")
            return
    # Si el bucle termina sin encontrar el código
    print(f"\nERROR: No se encontró ningún alumno con el código {codigo_a_modificar}.")

# Ejecución de prueba
modificar_calificacion(INVENTARIO_ALUMNOS)        
```

## Celda 5: Cálculo de Totales y Reporte <br>
Esta función (reporte_materias) agrupa los datos para realizar un cálculo o resumen por una categoría (simula el cálculo por unidad de medida o tipo de producto).

```python
# 5. Función para reporte de materias (Simula 'Calcular existencias totales por unidad/tipo')

def reporte_materias(inventario):
    """
    Calcula y muestra el promedio de calificación por materia registrada.
    (Simular a calcular el total de existencias por unidad de medida).
    """
    print("\n--- REPORTE DE CALIFICACIONES POR MATERIA ---")

    # Diccionario para almacenar la suma de calificaciones y el conteo por materia
    reporte_agregado = {} # { 'Materia_A': {'suma': 0, 'conteo': 0}, ... }

    # Recorrido y Agregación (Concepto clave para analisis)
    for alumno in inventario:
        materia = alumno['materia']
        calificacion = alumno['calificacion']

        if materia not in reporte_agregado:
            reporte_agregado[materia] = {'suma': 0, 'conteo': 0}
        
        reporte_agregado[materia]['suma'] += calificacion
        reporte_agregado[materia]['conteo'] += 1

    # Cálculo y presentación de promedios
    if not reporte_agregado:
        print("El inventario está vacío. No hay datos para reportar.")
        return
    
    for materia, datos in reporte_agregado.items():
        promedio = datos['suma'] / datos['conteo']
        print(f"- {materia}: {datos['conteo']} alumnos | Promedio: {promedio:.2f}")

# Ejecución de prueba
reporte_materias(INVENTARIO_ALUMNOS)
```

## Celda 6: Menú Interactivo y Modularidad <br>
Esta celda une todas las funciones anteriores en un menú interactivo, demostrando cómo se organiza el código final en funciones reutilizables.


```python
# 6. Función del Menú Principal (Organizar el código en funciones reutilizables)

def mostrar_menu():
    """Muestra las opciones del menú principal."""
    print("\n===========================================")
    print("    SISTEMA DE CONTROL DE CALIFICACIONES     ")
    print("===========================================")
    print("1. Registrar nuevo alumno")
    print("2. Consultar alumno (por código/nombre)")
    print("3. Modificar calificación")
    print("4. Reporte por materia (Existencias totales)")
    print("5. Mostrar todos los alumnos (Inventario completo)")
    print("6. Salir")
    print("===========================================")    

def mostrar_inventario_completo(inventario):
    """Simula mostrar el inventario completo."""
    if not inventario:
        print("\nEl inventario de alumnos está vacío.")
        return
    
    print("\n--- INVENTARIO COMPLETO DE ALUMNOS ---")
    for alumno in inventario:
        print(f"Cod: {alumno['codigo']} | Nombre: {alumno['nombre']:<20} | Calificación: {alumno['calificacion']:.2f} | Materia: {alumno['materia']}") 

def sistema_principal (inventario):
    """Función principal que ejecuta el menú interactivo."""

    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción (1-6): ").strip()
        
        if opcion == '1':
            registra_alumno(inventario)
        elif opcion == '2':
            criterio = input("Ingrese el código o parte del nombre a buscar: ")
            resultados = buscar_alumno(inventario, criterio)
            if resultados:
                for a in resultados:
                    print(f"ENCONTRADO -> Cód: {a['codigo']}, Nombre: {a['nombre']}, Calificación: {a['calificacion']}")
            else:
                print("Búsqueda sin resultados.")
        elif opcion == '3':
            modificar_calificacion(inventario)
        elif opcion == '4':
            reporte_materias(inventario)
        elif opcion == '5':
            mostrar_inventario_completo(inventario)
        elif opcion == '6':
            print("Saliendo del sistema. ¡Hasta pronto!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Para ejecutar el sistema, descomente la siguiente línea:
sistema_principal(INVENTARIO_ALUMNOS)
```
